```python
mytrain=[]

for k in np.arange(0,len(trainset)):
    #the ground truth
    ta=np.copy(trainset[k][0])
    
    #Downsample and upscale the image again
    todown=np.copy(trainset[k][0])
    todown=np.transpose(todown, (1, 2, 0))

    temp=np.copy(todown)
    for i in np.arange(0,3):
        temp[:,:,i]=ndi.gaussian_filter(todown[:,:,i],sigma=1)

    myLR=np.zeros((8,8,3),dtype=np.float32)

    for i in np.arange(2,32,4):
        for j in np.arange(2,32,4):
            myLR[int((i-2)/4),int((j-2)/4),:]=(temp[i,j,:])
    
    mybicu=bicubic.bicubic(myLR,4,-1/2,k)
    mybicu=np.transpose(mybicu, (2,0,1))
    
    q=np.where(mybicu<0)
    mybicu[q]=0.
    q=np.where(mybicu>1)
    mybicu[q]=1.
    
    
    ta=torch.from_numpy(ta)
    final=torch.from_numpy(mybicu).float()
    
    mytrain.append([final,ta])
```
